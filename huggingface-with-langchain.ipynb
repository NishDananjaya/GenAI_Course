**text2textgeneration models | seq2seq models | Encoder decoder models**

```python
prompt = PromptTemplate(
    input_variables = ['product'],
    template = 'What is a good name for a company that makes {product}'
)
```

# **Text generation models only | Decoder only models**

```python
from langchain.llms import HuggingFacePipeline
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, AutoModelForSeq2SeqLM
```

```python
prompt = PromptTemplate(
    input_variables = ['name'],
    template = 'Can you tell me about footballer {name}'
)
```

```python
model_id = 'google/flan-t5-large'
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained(model_id,device_map = 'auto')
```

```python
pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer, max_length=128)
```

```python
local_llm = HuggingFacePipeline(pipeline = pipeline)
```

```python
prompt = PromptTemplate(
    input_variables = ['name'],
    template = 'Can you tell me about footballer {name}'
)
```

```python
chain = LLMChain(llm = local_llm,prompt = prompt)
```

```python
chain.run("Messi")
```

```python
prompt2 = PromptTemplate(
    input_variables = ['name'],
    template = 'Can you tell me about cricketer {name}'
)
```

```python
chain = LLMChain(llm = local_llm,prompt = prompt2)
```

```python
chain.run("Kumar Sangakkara")
```
